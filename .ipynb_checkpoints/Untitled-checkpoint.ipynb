```python
n =int(input("Enter a value"))
sum = 0

while(n!=0):
    digit = n%10
    sum = sum + digit
    n//=10
print(f"The sum of the digits is : {sum} ")
```

```python
n = int(input("Enter a value"))
rev = 0
while(n!=0):
    digit = n%10
    rev = rev*10 + digit
    n//=10
print(f"The Reverse of the digit is {rev} ")
```

```python
def palin(n):
    rev = 0
    temp = n
    while(n!=0):
        digit = n%10
        rev = rev*10 + digit
        n//=10
    if(rev == temp):
        print(f"This is a palindrome No. {rev} ")
    else:
        print(f"This is not a palindrome no.  ")

palin(1373)
```

```python
def arm(n):
    temp = n
    rev = 0
    while(n!=0):
        digit = n%10
        rev = digit**3 + rev
        n//=10
    if(rev == temp):
        print(f"This is an armstrong no. {rev} ")
    else:
        print(f"This is not an armstrong no.")

arm(371)
```
